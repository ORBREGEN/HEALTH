```python
# ================================================
# 01_load_hlca_full.ipynb
# Safe loading of HLCA Full (20.36 GB)
# ================================================

# Install packages in Colab environment
%pip install scanpy anndata pandas numpy tqdm --quiet

print("✅ Packages installed successfully!")

import scanpy as sc
import os
from tqdm import tqdm
import requests

print("✅ Packages installed!\n")

# ====================== DOWNLOAD & LOAD HLCA FULL ======================
full_url = "https://datasets.cellxgene.cziscience.com/dbb5ad81-1713-4aee-8257-396fbabe7c6e.h5ad"
local_path = "data/hlca_full.h5ad"

os.makedirs("data", exist_ok=True)

if not os.path.exists(local_path):
    print("Downloading 20.36 GB file... (this may take 15-40 minutes)")
    response = requests.get(full_url, stream=True)
    total_size = int(response.headers.get('content-length', 0))
    
    with open(local_path, "wb") as f, tqdm(
        desc="Downloading", total=total_size, unit='iB', unit_scale=True, unit_divisor=1024
    ) as bar:
        for data in response.iter_content(chunk_size=1024*1024):
            size = f.write(data)
            bar.update(size)
    print("✅ Download finished!")
else:
    print("✅ File already exists.")

# Load in backed mode (safe for large files)
print("\nLoading dataset...")
adata = sc.read_h5ad(local_path, backed='r')

print("✅ SUCCESS! HLCA Full is loaded.")
print(f"Cells: {adata.shape[0]:,}")
print(f"Genes: {adata.shape[1]:,}")
```

```python
%pip install scanpy anndata pandas numpy tqdm requests --quiet
print("✅ Packages installed!")

import scanpy as sc
import os
from tqdm import tqdm
import requests
import time

url = "https://datasets.cellxgene.cziscience.com/a13eb499-f3ae-4c9b-b9ce-89fbddd17d53.h5ad"
save_path = "data/lung_tissues_visium_parenchyma.h5ad"

print("Downloading Visium Lung Parenchyma sample (~831 MB)...")

os.makedirs("data", exist_ok=True)

def download_with_resume(url, path, retries=5):
    for attempt in range(retries):
        try:
            headers = {}
            mode = "wb"

            # Resume support
            if os.path.exists(path):
                existing_size = os.path.getsize(path)
                headers["Range"] = f"bytes={existing_size}-"
                mode = "ab"
                print(f"🔁 Resuming from {existing_size / (1024*1024):.2f} MB")

            response = requests.get(url, headers=headers, stream=True, timeout=30)
            response.raise_for_status()

            total_size = int(response.headers.get("content-length", 0))
            if "Range" in headers:
                total_size += existing_size

            with open(path, mode) as f, tqdm(
                desc="Downloading",
                total=total_size,
                unit='iB',
                unit_scale=True,
                unit_divisor=1024
            ) as bar:

                if "Range" in headers:
                    bar.update(existing_size)

                for chunk in response.iter_content(chunk_size=1024 * 1024):
                    if chunk:
                        f.write(chunk)
                        bar.update(len(chunk))

            return True

        except Exception as e:
            print(f"❌ Attempt {attempt+1} failed: {e}")
            time.sleep(5)

    return False


# Download if needed
if not os.path.exists(save_path) or os.path.getsize(save_path) < 800 * 1024 * 1024:
    success = download_with_resume(url, save_path)

    if not success:
        raise Exception("Download failed after multiple attempts")
else:
    print("✅ File already exists and looks complete.")


# Validate file size before loading
file_size = os.path.getsize(save_path)
print(f"📦 File size: {file_size / (1024*1024):.2f} MB")

if file_size < 800 * 1024 * 1024:
    raise Exception("❌ File is incomplete. Please re-download.")


# Load safely
print("\nLoading Visium Parenchyma dataset...")
adata_visium_parenchyma = sc.read_h5ad(save_path, backed='r')

print("✅ Visium Lung Parenchyma loaded successfully!")
print(f"Cells/Spots : {adata_visium_parenchyma.shape[0]:,}")
print(f"Genes       : {adata_visium_parenchyma.shape[1]:,}")
```

```python
# ================================================
# Download + Extract HLCA Reference Model (.zip)
# ================================================

%pip install scanpy anndata pandas numpy tqdm requests --quiet

import os
from tqdm import tqdm
import requests
import zipfile

url = "https://zenodo.org/records/7599104/files/HLCA_reference_model.zip?download=1"
save_path = "data/HLCA_reference_model.zip"
extract_path = "data/HLCA_reference_model_extracted"

print("Downloading HLCA Reference Model (.zip)...")

os.makedirs("data", exist_ok=True)

if not os.path.exists(save_path):
    response = requests.get(url, stream=True)
    total_size = int(response.headers.get('content-length', 0))
    
    with open(save_path, "wb") as f, tqdm(
        desc="Downloading", 
        total=total_size, 
        unit='iB', 
        unit_scale=True, 
        unit_divisor=1024
    ) as bar:
        for data in response.iter_content(chunk_size=1024*1024):
            size = f.write(data)
            bar.update(size)
    print("✅ Download completed!")
else:
    print("✅ File already exists.")

# Extract the zip file
print("\nExtracting the zip file...")
if not os.path.exists(extract_path):
    with zipfile.ZipFile(save_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    print("✅ Extraction completed!")
else:
    print("✅ Already extracted.")

print(f"\nExtracted folder: {extract_path}")
print("You can now explore the files inside it.")
```
